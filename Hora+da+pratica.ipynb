````python

# pede os dois n√∫meros
num1 = float(input('Digite o primeiro n√∫mero: '))
num2 = float(input('Digite o segundo n√∫mero: '))

# pergunta a opera√ß√£o
operacao = input('Qual opera√ß√£o deseja realizar? (+, -, *, /): ')

# decide qual conta fazer
if operacao == '+':
    resultado = num1 + num2
elif operacao == '-':
    resultado = num1 - num2
elif operacao == '*':
    resultado = num1 * num2
elif operacao == '/':
    resultado = num1 / num2
else:
    print('Opera√ß√£o inv√°lida!')
    resultado = None  # importante pra evitar erro depois

# s√≥ continua se a opera√ß√£o for v√°lida
if resultado is not None:

    print(f'Resultado: {resultado}')

    # üëâ inteiro ou decimal
    if resultado == int(resultado):
        tipo = 'inteiro'
    else:
        tipo = 'decimal'

    # üëâ positivo, negativo ou zero
    if resultado > 0:
        sinal = 'positivo'
    elif resultado < 0:
        sinal = 'negativo'
    else:
        sinal = 'zero'

    # üëâ par ou √≠mpar (s√≥ faz sentido se for inteiro)
    if tipo == 'inteiro':
        if int(resultado) % 2 == 0:
            paridade = 'par'
        else:
            paridade = '√≠mpar'
    else:
        paridade = 'n√£o se aplica'

    print(f'O n√∫mero √© {tipo}, {sinal} e {paridade}.')

````

#### Quest√£o 12
Um estabelecimento est√° vendendo combust√≠veis com descontos variados. Para o etanol, se a quantidade comprada for at√© 15 litros, o desconto ser√° de 2% por litro. Caso contr√°rio, ser√° de 4% por litro. Para o diesel, se a quantidade comprada for at√© 15 litros, o desconto ser√° de 3% por litro. Caso contr√°rio, ser√° de 5% por litro. O pre√ßo do litro de diesel √© R\$ 2,00 e o pre√ßo do litro de etanol √© R\$ 1,70. Escreva um programa que leia a quantidade de litros vendidos e o tipo de combust√≠vel (E para etanol e D para diesel) e calcule o valor a ser pago pelo cliente. Tenha em mente algumas dicas:

- O do valor do desconto ser√° a multiplica√ß√£o entre pre√ßo do litro, quantidade de litros e o valor do desconto.
- O valor a ser pago por um cliente ser√° o resultado da multiplica√ß√£o do pre√ßo do litro pela quantidade de litros menos o valor de desconto resultante do c√°lculo.


````python
print("-" * 30)
print("  POSTO DE COMBUST√çVEL F1  ")
print("-" * 30)

litros = float(input("Quantos litros voc√™ quer abastecer? "))
tipo = input("Qual o combust√≠vel? (E para Etanol / D para Diesel): ").upper()

# Definindo pre√ßos base
preco_etanol = 1.70
preco_diesel = 2.00

# L√ìGICA DO ETANOL
if tipo == 'E':
    preco_total = litros * preco_etanol
    if litros <= 15:
        desconto = preco_total * 0.02 # 2% de desconto
    else:
        desconto = preco_total * 0.04 # 4% de desconto

# L√ìGICA DO DIESEL
elif tipo == 'D':
    preco_total = litros * preco_diesel
    if litros <= 15:
        desconto = preco_total * 0.03 # 3% de desconto
    else:
        desconto = preco_total * 0.05 # 5% de desconto

else:
    print("Tipo de combust√≠vel inv√°lido!")
    # Criamos uma vari√°vel de controle para n√£o dar erro no print final
    preco_total = 0
    desconto = 0

# C√°lculo final
valor_final = preco_total - desconto

if valor_final > 0:
    print("-" * 30)
    print(f"Combust√≠vel selecionado: {tipo}")
    print(f"Valor bruto: R$ {preco_total:.2f}")
    print(f"Desconto aplicado: R$ {desconto:.2f}")
    print(f"TOTAL A PAGAR: R$ {valor_final:.2f}")
    print("-" * 30)
````

Tendo esses valores, fa√ßa um c√≥digo que gere uma lista contendo o percentual de crescimento de bact√©rias por dia, comparando o n√∫mero de bact√©rias em cada dia com o n√∫mero de bact√©rias do dia anterior. Dica: para calcular o percentual de crescimento usamos a seguinte equa√ß√£o: `100 * (amostra_atual - amostra_passada) / (amostra_passada)`.

````python
bacterias = [1.2, 2.1, 3.3, 5.0, 7.8, 11.3, 16.6, 25.1, 37.8, 56.9]

crescimento_percentual = []

for i in range(1, len(bacterias)):
    amostra_passada = bacterias[i - 1]
    amostra_atual = bacterias[i]

    crescimento = 100 * (amostra_atual - amostra_passada) / amostra_passada

    crescimento_percentual.append(crescimento)

print("Percentual de crescimento di√°rio:")
print(crescimento_percentual)
````

Uma equipe de cientistas de dados est√° estudando a diversidade biol√≥gica em uma floresta. A equipe fez a coleta de informa√ß√µes sobre o n√∫mero de esp√©cies de plantas e animais em cada √°rea da floresta e armazenaram essas informa√ß√µes em um dicion√°rio. Nele, a chave descreve a √°rea dos dados e os valores nas listas correspondem √†s esp√©cies de plantas e animais nas √°reas, respectivamente.
```
{'√Årea Norte': [2819, 7236],
 '√Årea Leste': [1440, 9492],
 '√Årea Sul': [5969, 7496],
 '√Årea Oeste': [14446, 49688],
 '√Årea Centro': [22558, 45148]}
```
Escreva um c√≥digo para calcular a m√©dia de esp√©cies por √°rea e identificar a √°rea com a maior diversidade biol√≥gica. Dica: use as fun√ß√µes built-in [`sum()`](https://docs.python.org/3/library/functions.html#sum) e [`len()`](https://docs.python.org/3/library/functions.html#len).


````python
dados = {
    '√Årea Norte': [2819, 7236],
    '√Årea Leste': [1440, 9492],
    '√Årea Sul': [5969, 7496],
    '√Årea Oeste': [14446, 49688],
    '√Årea Centro': [22558, 45148]
}

medias = {}

maior_diversidade = 0
area_maior = ""

# calcular m√©dia e diversidade por √°rea
for area, especies in dados.items():
    total = sum(especies)              # soma plantas + animais
    media = total / len(especies)      # m√©dia das esp√©cies na √°rea

    medias[area] = media

    # verificar maior diversidade (maior total)
    if total > maior_diversidade:
        maior_diversidade = total
        area_maior = area

# resultados
print("--- M√©dia de esp√©cies por √°rea ---")
for area, media in medias.items():
    print(f"{area}: m√©dia = {media:.2f}")

print("\n--- √Årea com maior diversidade biol√≥gica ---")
print("√Årea:", area_maior)
print("Total de esp√©cies:", maior_diversidade)
````

#### Quest√£o 15
O setor de RH da sua empresa te pediu uma ajuda para analisar as idades dos funcion√°rios de 4 setores da empresa. Para isso, ele te forneceu os seguintes dados:
```
{'Setor A': [22, 26, 30, 30, 35, 38, 40, 56, 57, 65],
 'Setor B': [22, 24, 26, 33, 41, 49, 50, 54, 60, 64],
 'Setor C': [23, 26, 26, 29, 34, 35, 36, 41, 52, 56],
 'Setor D': [19, 20, 25, 27, 34, 39, 42, 44, 50, 65]}
```
Sabendo que cada setor tem 10 funcion√°rios, construa um c√≥digo que calcule a m√©dia de idade de cada setor, a idade m√©dia geral entre todos os setores e quantas pessoas est√£o acima da idade m√©dia geral.


````python
idades = {
    'Setor A': [22, 26, 30, 30, 35, 38, 40, 56, 57, 65],
    'Setor B': [22, 24, 26, 33, 41, 49, 50, 54, 60, 64],
    'Setor C': [23, 26, 26, 29, 34, 35, 36, 41, 52, 56],
    'Setor D': [19, 20, 25, 27, 34, 39, 42, 44, 50, 65]
}

medias_setores = {}
todas_idades = []

# 1. M√©dia de cada setor
for setor, lista_idades in idades.items():
    media = sum(lista_idades) / len(lista_idades)
    medias_setores[setor] = media

    # juntar todas as idades para calcular m√©dia geral depois
    todas_idades.extend(lista_idades)

# 2. M√©dia geral
media_geral = sum(todas_idades) / len(todas_idades)

# 3. Quantas pessoas est√£o acima da m√©dia geral
acima_media = 0

for idade in todas_idades:
    if idade > media_geral:
        acima_media += 1

# ŸÜÿ™ÿßÿ¶ÿ¨
print("--- M√©dia de idade por setor ---")
for setor, media in medias_setores.items():
    print(f"{setor}: {media:.2f} anos")

print("\n--- M√©dia geral ---")
print(f"M√©dia geral de idade: {media_geral:.2f} anos")

print("\n--- Pessoas acima da m√©dia geral ---")
print("Quantidade:", acima_media)
````
